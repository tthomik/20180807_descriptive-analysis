```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from mpl_toolkits import mplot3d
```

```python
#Datenset eingelesen
data = pd.read_csv("data/zip.train", header=None, sep=" ")
```

```python
cleaned_data = data.iloc[:,:-1]
```

```python
def number_figure(n):
    number = pd.DataFrame(cleaned_data[cleaned_data[0] == n].iloc[:,1:])
    number_median = number.median()
    image_reshape = number_median.values.reshape((16,16))
    plt.imshow(image_reshape, cmap="magma", interpolation="none")
    return plt.show()

numbers = range(0,10)

for n in numbers:
    number_figure(n)
```

```python
def number_median(n):
    number = pd.DataFrame(cleaned_data[cleaned_data[0] == n].iloc[:,1:])
    return number.median()

numbers = range(0,10)

n0 = number_median(0)
n1 = number_median(1)
n2 = number_median(2)
n3 = number_median(3)
n4 = number_median(4)
n5 = number_median(5)
n6 = number_median(6)
n7 = number_median(7)
n8 = number_median(8)
n9 = number_median(9)
```

```python
y_werte = range(0,256)
median = [n0,n1,n2,n3,n4,n5,n6,n7,n8,n9]

def scatter_plots(n):
    plt.scatter(y_werte,n)
    return plt.show()
    
for n in median:
    scatter_plots(n)
```

```python
superclean_data = cleaned_data.iloc[:,1:]

mixture = GaussianMixture(n_components=10, covariance_type='full', n_init=10).fit(superclean_data)
cleaned_data['mixture'] = mixture.predict(superclean_data)
```

```python
cleaned_data.head()
```

```python
plt.scatter(cleaned_data[0],cleaned_data['mixture'], c=cleaned_data['mixture'])
plt.xlabel = "Zahlen"
plt.ylabel = "Cluster"
plt.show()
```

```python
def number_df(n):
    number = pd.DataFrame(cleaned_data[cleaned_data[0] == n].iloc[:,1:])
    return pd.DataFrame(number)

numbers = range(0,10)

df0 = number_df(0)
df1 = number_df(1)
df2 = number_df(2)
df3 = number_df(3)
df4 = number_df(4)
df5 = number_df(5)
df6 = number_df(6)
df7 = number_df(7)
df8 = number_df(8)
df9 = number_df(9)

def count_row(n):
    return len(n.axes[0])

row0 = count_row(df0)
row1 = count_row(df1)
row2 = count_row(df2)
row3 = count_row(df3)
row4 = count_row(df4)
row5 = count_row(df5)
row6 = count_row(df6)
row7 = count_row(df7)
row8 = count_row(df8)
row9 = count_row(df9)
```

```python
z_werte = range(0,7291)

fig = plt.figure(figsize=(20,8))
ax = plt.axes(projection = '3d')
ax.scatter3D(cleaned_data[0],cleaned_data['mixture'], z_werte, c=cleaned_data['mixture'])
ax.set_xlabel('Ziffer')
ax.set_ylabel('Cluster-Nummer')
ax.set_zlabel('Anzahl der Ziffern in einem Cluster')
```

```python
cleaned_data[cleaned_data[0] == 0].groupby("mixture").count()[0]
```

```python
bar_data = np.zeros((10,10))
for digit in range(0,10):
    temp_result = cleaned_data[cleaned_data[0] == digit].groupby('mixture').count()[0]
    cluster_keys = temp_result.keys()
    for cluster_idx in cluster_keys:
        bar_data[digit][cluster_idx] = temp_result[cluster_idx]
```

```python
result_df = pd.DataFrame(bar_data)
```
