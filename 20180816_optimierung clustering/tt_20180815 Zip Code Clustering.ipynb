```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.mixture import GaussianMixture
from mpl_toolkits import mplot3d
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

# Clustering mit Gaussian

```python
#Datenset eingelesen
data = pd.read_csv("zip.train", header=None, sep=" ")
```

```python
cleaned_data = data.iloc[:,:-1]
```

```python
#den Median über alle einzelnen Zahlen jeder Ziffer als Bild darstellen

def number_figure(n):
    number = pd.DataFrame(cleaned_data[cleaned_data[0] == n].iloc[:,1:])
    number_median = number.median()
    image_reshape = number_median.values.reshape((16,16))
    plt.imshow(image_reshape, cmap=plt.cm.binary, interpolation="none")
    return plt.show()

numbers = range(0,10)

for n in numbers:
    number_figure(n)
```

```python
#Berechnung der Mediane über alle Zahlen jeder Ziffer

def number_median(n):
    number = pd.DataFrame(cleaned_data[cleaned_data[0] == n].iloc[:,1:])
    return number.median()

mediane = []

for number in range(0,10):
    mediane.append(number_median(number))
```

```python
#Darstellung der Mediane der Zahlen jeder Ziffer in Form von Dotplots mit den Pixel auf der x-Achse

x_werte = range(0,256)

def scatter_plots(n):
    plt.scatter(x_werte,n)
    return plt.show()
    
for median in mediane:
    scatter_plots(median)
```

```python
#Entfernung der ersten Spalte, in der die dargestellte Zahl steht
superclean_data = cleaned_data.iloc[:,1:]

#Berechnung der Gaussian Mixture aller Zahlen
mixture = GaussianMixture(n_components=10, covariance_type='full', n_init=1, random_state=1).fit(superclean_data)
```

```python
#Kopieren von cleaned_data, damit das df unverändert bleibt
cleaned_data_mixture = pd.DataFrame.copy(cleaned_data,deep=True)

#Anfügen der Cluster-Wahrscheinlichkeiten (pd_mixture_probs) und des ausgewählten Clusters ('mixture') an ursrpüngliche Daten
mixture_probs = pd.DataFrame(mixture.predict_proba(superclean_data))
pd_mixture_probs = mixture_probs.rename(columns ={0: 'Cluster 0', 1: 'Cluster 1', 2: 'Cluster 2', 3: 'Cluster 3', 4: 'Cluster 4', 5: 'Cluster 5', 6: 'Cluster 6', 7: 'Cluster 7', 8: 'Cluster 8', 9: 'Cluster 9'})
cleaned_data_results = pd.concat((cleaned_data_mixture,pd_mixture_probs), axis=1)
cleaned_data_results['mixture'] = mixture.predict(superclean_data)
cleaned_data_mixture['mixture'] = mixture.predict(superclean_data)

cleaned_data_results.head()
```

```python
#erstellen eines dfs, das die dargestellte Zahl, die Cluster-Wahrscheinlichkeiten und die ausgewählten Cluster enthält
mixture_numbers = pd.DataFrame()
mixture_numbers['mixture'] = mixture.predict(superclean_data)
mixture_numbers = pd.DataFrame(mixture_numbers)
mixture_results_overview = pd.concat((cleaned_data_mixture[0],pd_mixture_probs, mixture_numbers), axis=1)
```

```python
mixture_results_overview.head()
```

```python
#Scatter-Plot, Zahlen gegen Cluster aufgetragen, Dotgröße je nach Wahrscheinlichkeit dargestellt

size = 50 * mixture_probs.max(1) ** 2

plt.scatter(cleaned_data[0],cleaned_data_mixture['mixture'], s=size, c=cleaned_data_mixture['mixture'])
plt.xlabel("Zahlen")
plt.ylabel("Cluster")
plt.show()
```

```python
#3D-Blot, z-Achse die Anzahl aller einzelnen Zahlen, Dotgröße nach Wahrscheinlichkeit

z_werte = range(0,7291)

fig = plt.figure(figsize=(20,10))
ax = plt.axes(projection = '3d')
ax.scatter3D(cleaned_data[0],cleaned_data_mixture['mixture'], z_werte, s=size, c=cleaned_data_mixture['mixture'])
ax.set_xlabel('Ziffer')
ax.set_ylabel('Cluster-Nummer')
ax.set_zlabel('Anzahl der Ziffern in einem Cluster')
plt.show()
```

```python
#ziehe aus dem df cleaned_data alle Zeilen, die in der 1. Spalte (0) die Zahl 0 stehen haben und grupiere die Zeilen
#nach den zugeordneten Clustern ('mixture'), dann zähle, wie viele Zeilen es zu jedem Cluster gibt und gebe das in einem
#df raus, gebe vom df nur die erste Spalte raus, da die restlichen nicht von Bedeutung sind
cleaned_data_mixture[cleaned_data_mixture[0] == 0].groupby("mixture").count()[0].keys()
```

```python
bar_data = np.zeros((10,10)) #lege 10/10 Matrix an, mit 0 gefüllt
for digit in range(0,10): #für jede Zahl zwischen 0 und 10
    temp_result = cleaned_data_mixture[cleaned_data_mixture[0] == digit].groupby('mixture').count()[0]
    cluster_keys = temp_result.keys() #gibt mir den Index (erste Spalte) für temp_results als Liste raus
    for cluster_idx in cluster_keys: #für jeden Cluster in der Cluster-Sammlung (cluster_keys)...
        bar_data[digit][cluster_idx] = temp_result[cluster_idx]
```

```python
#Automatisches Hinzufügen der Spaltennamen zu bar_data
col_names = []
for cluster_idx in range(0,10):
    col_names.append('Cluster_' + str(cluster_idx))
print (col_names)
result_df = pd.DataFrame(bar_data, columns=col_names)
```

```python
result_df
```

```python
t_result_df = result_df.transpose()
t_result_df
```

```python
#Balkendiagramm der Aufteilung der Cluster über die Zahlen
result_df.plot.bar(figsize=(20,8))
plt.xlabel("Ziffer")
plt.ylabel("Häufigkeit")
plt.show()
```

```python
#Normierung über die Summe jeder Zahl
normiert = t_result_df / t_result_df.sum(0)
```

```python
normiert
```

```python
#Normiertes Balkendiagramm, Aufteilung der Cluster auf die Zahlen
normiert.T.plot.bar(figsize=(20,8))
plt.title("normiert")
plt.xlabel("Ziffer")
plt.ylabel("Relative Häufigkeit")
plt.show()
```

# Clustering nach KMeans

```python
#Durchfühurg von KMeans
cleaned_data_kmeans = pd.DataFrame.copy(cleaned_data, deep=True)
kmeans = KMeans(n_clusters=10,n_init=100).fit(superclean_data)
cleaned_data_kmeans['cluster_nr'] = kmeans.predict(superclean_data)
cleaned_data_kmeans.head()
```

```python
#Zusammenfassung der Ergebnisse: auf wie viele der einzelnen Zahlen jeder Ziffer sind in welchen Clustern
bar_data_kmeans = np.zeros((10,10))
for digit in range(0,10):
    temp_result_kmeans = cleaned_data_kmeans[cleaned_data_kmeans[0] == digit].groupby('cluster_nr').count()[0]
    cluster_keys_kmeans = temp_result_kmeans.keys() #gibt mir den Index (erste Spalte) für temp_results als Liste raus
    for cluster_idx in cluster_keys_kmeans: #für jeden Cluster in der Cluster-Sammlung (cluster_keys)...
        bar_data_kmeans[digit][cluster_idx] = temp_result_kmeans[cluster_idx]
```

```python
col_names_kmeans = []
for cluster_idx in range(0,10):
    col_names_kmeans.append('Cluster_' + str(cluster_idx))
print (col_names_kmeans)
result_kmeans_df = pd.DataFrame(bar_data_kmeans, columns=col_names_kmeans)
```

```python
result_kmeans_df
```

```python
t_result_kmeans_df = result_kmeans_df.transpose()
t_result_kmeans_df
```

```python
result_kmeans_df.plot.bar(figsize=(20,8))
plt.xlabel("Ziffer")
plt.ylabel("Häufigkeit")
plt.show()
```

```python
normiert_kmeans = result_kmeans_df.divide(result_kmeans_df.sum(1), axis=0)
```

```python
normiert_kmeans.plot.bar(figsize=(20,8))
plt.title("normiert")
plt.xlabel("Ziffer")
plt.ylabel("Relative Häufigkeit")
plt.show()
```

# Vorbearbeitung der Daten mit t-SNE 

```python
cleaned_data_kmeans_tsne = pd.DataFrame.copy(cleaned_data, deep=True)
```

```python
tsne = TSNE(n_components=2, init='random', random_state=0)
digits_proj = tsne.fit_transform(superclean_data)

kmeans_tsne = KMeans(n_clusters=10, random_state=0)
clusters_tsne = kmeans_tsne.fit_predict(digits_proj)
```

```python
cleaned_data_kmeans_tsne['cluster_nr'] = pd.Series(clusters_tsne, index=cleaned_data_kmeans_tsne.index)
cleaned_data_kmeans_tsne.head()
```

```python
bar_data_kmeans_tsne = np.zeros((10,10))
for digit in range(0,10):
    temp_result_kmeans_tsne = cleaned_data_kmeans_tsne[cleaned_data_kmeans_tsne[0] == digit].groupby('cluster_nr').count()[0]
    cluster_keys_kmeans_tsne = temp_result_kmeans_tsne.keys() #gibt mir den Index (erste Spalte) für temp_results als Liste raus
    for cluster_idx in cluster_keys_kmeans_tsne: #für jeden Cluster in der Cluster-Sammlung (cluster_keys)...
        bar_data_kmeans_tsne[digit][cluster_idx] = temp_result_kmeans_tsne[cluster_idx]

col_names_kmeans_tsne = []
for cluster_idx in range(0,10):
    col_names_kmeans_tsne.append('Cluster_' + str(cluster_idx))
print (col_names_kmeans_tsne)
result_kmeans_tsne_df = pd.DataFrame(bar_data_kmeans_tsne, columns=col_names_kmeans_tsne)
result_kmeans_tsne_df
```
