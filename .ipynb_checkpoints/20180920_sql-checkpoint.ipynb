### Gruppieren von Daten

```python
# Gruppiere die Daten aus 'answers' anhand der Einträge in colorname.
# Wähle nur die Eintrage welche die Buchstaben 'blue' enhalten aus (WHERE).
# Gebe den Eintrag 'colorname' aus und zähle (COUNT(*)) die Anzahl der Einträge pro Gruppe 
# und gebe diesen ebenfalls aus. Weise diesem Wert den Namen 'numer_entries' (AS) zu.
# Gebe nur die Gruppen aus die mehr als 100 Einträge (HAVING) haben aus
# Sortiere das Ergebnis nach der Anzahl der Einträge (ORDER BY) absteigend (DESC)
pd.read_sql_query(
    'SELECT ' +
        'colorname, ' +
        'COUNT(*) AS number_entries ' +
    'FROM answers ' +
    'WHERE colorname LIKE "%blue%" ' +
    'GROUP BY colorname ' +
    'HAVING number_entries > 100 ' +
    'ORDER BY number_entries DESC; '
    , conn)
```

### Daten aus verschiedenen Tabellen joinen

Möglichkeiten Einträge zweier Tabellen zusammenzuführen, funktioniert mittels den 'JOIN' Befehlen. Zwei oder mehr Tabellen werden auf (ON) einer Bedingung zusammengeführt.

'LEFT JOIN' (alle Datenreien der linken Tabelle sind im Ergebnis enthalten, auch wenn es keine Daten dazu in der rechten Tabelle gibt)

'RIGHT JOIN' (alle Datenreien der rechten Tabelle sind im Ergebnis enthalten, auch wenn es keine Daten dazu in der linken Tabelle gibt) 

'OUTER JOIN' (alle Datenreien beider Tabellen sind im Ergebnis enthalten)
'INNER JOIN' (nur die Datenreien zu denen es Daten in beiden Tabellen gibt sind enthalten)

```python
df = pd.read_sql_query('SELECT * FROM answers INNER JOIN users ON users.id = answers.user_id;', conn)
```

### Herausfinden, ab welcher Häufigkeit noch "sinnvolle" Namen für Farben auftauchen.
- Für jede dieser Farben dann einen Farbraum für r, g und b angeben
- wie kann ich für jeden colorname mir alle dazugehörigen r-, g- und b-Werte ausgeben?
- Für jeden Eintrag in Spalte colorname in colors soll df4 durchsucht werden und jede dazugehörige r,g,b Werte soll in ein neues df geschrieben werden, damit min- und max-Werte von jeder Farbe ausgelesen werden können.
- hierdurch können alle Farben zu einem Farbeintrag dargestellt werden. Wäre der Mittelwert jedes Wert dann die mittlere Farbe?

```python
# Auswahl eines Treshholds, um sinnlose Farbnamen auszuschließen. Treshold = Häufigkeit einer Farbe.
# mit diesem Treshold (4554) habe ich 76 verschiedene Farben
colors = pd.read_sql_query(
    'SELECT ' +
        'colorname, ' +
        'COUNT(*) AS number_entries ' +
    'FROM answers ' +
    'WHERE colorname LIKE "%%" ' +
    'GROUP BY colorname ' +
    'HAVING number_entries > 4554 ' +
    'ORDER BY number_entries Asc; '
    , conn)
colors.tail(100)
```

```python
colorlist = list(map(lambda x:f'"{x}"', colors["colorname"].tolist()))
colorlist
```

```python
sql = ('SELECT ' +
'colorname,r,g,b ' +
    'FROM answers ' +
    'WHERE colorname IN (%s) ' +
    'ORDER BY colorname Asc; ') % (','.join(list(colorlist)))
sql
```

```python
colors_rgb = pd.read_sql_query(sql, conn)
colors_rgb.head()
```
