# Verwendung der logistischen Regression für Zifferndatensatz

### import der benötigten libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from scipy.stats import mode
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.manifold import TSNE
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from scipy.ndimage.interpolation import shift
```

### Laden des Datenset

```python
#Datensets einlesen
digit_train = pd.read_csv("Data/zip.train", header=None, sep=" ") # training set
digit_test = pd.read_csv("Data/zip.test", header=None, sep=" ") # test set
```

```python
# Entfernung der letzten Spalte vom training set
digit_train_clean = digit_train.iloc[:,:-1]
digit_train_clean.head()
```

### Zuweisung der Daten für train, train_true, test, test_true

```python
# Entfernung der Spalte 0 (tatsächliche Ziffern), um trainieren zu können
x_train = digit_train_clean.drop(0, axis = 1)
x_train.head()
```

```python
len(x_train)
```

```python
# Definition des labels(y) jedes Datensatzes
y_train = digit_train_clean[0]
y_train.head()
```

```python
len(y_train)
```

```python
# Testdatensatz
x_test = digit_test.drop(0, axis = 1) # ohne true-Wert
y_test = digit_test[0] # Zielwert, true-Wert, label
```

## Fitten des Models logistische Regression und predicten der Ziffern für Trainingsdatensatz und Testdatensatz

```python
# Zuweisung der Funktion einer variablen
logisticRegr = LogisticRegression()
```

```python
# Trainieren des models mit den Daten
logisticRegr.fit(x_train, y_train)
```

```python
# Vorhersage der Ziffern des Trainingsdatensatzes
y_predict_train = logisticRegr.predict(x_train)
y_predict_train
```

```python
# Vorhersage der ziffern des Testdatensatzes
y_predict_test = logisticRegr.predict(x_test)
y_predict_test
```

### Berechnung der Accuracy und Fehler

```python
# Erstellung leerer Listen, damit numpy-Werte dort später angefügt werden können
acc_Train= []
acc_Test =[]
acc_Train_Error = []
acc_Test_Error = []

# Berechnung der accuracy für train und test
acc_train = accuracy_score(y_predict_train, y_train)
acc_test = accuracy_score(y_predict_test, y_test)

# Berechnung der Fehler: Fehler = 1 - acc
acc_train_error = 1 - acc_train
acc_test_error = 1 - acc_test

# Anfügen der Daten zu einer Liste
acc_Train.append(acc_train)
acc_Test.append(acc_test)
acc_Train_Error.append(acc_train_error)
acc_Test_Error.append(acc_test_error)
```

```python
# Ausgabe der berechneten accuracy werte in DataFrame
pd_accuracy_score_train_test = pd.DataFrame()
pd_accuracy_score_train_test["Train Accuracy"] = acc_Train
pd_accuracy_score_train_test["Test Accuracy"] = acc_Test
pd_accuracy_score_train_test["Train Error"] = acc_Train_Error
pd_accuracy_score_train_test["Test Error"] = acc_Test_Error
pd_accuracy_score_train_test
```

### Berechnung von Log Loss
#### Hierfür müssen zuerst die Wahrscheinlichkeiten für die Zuordnung zu einer Ziffer angegeben werden

```python
# Ausgabe der Wahrscheinlichkeit für Traindatensatz, dass die Pixelwerte einer Ziffer zugeordnet wurden
y_predict_train_proba = logisticRegr.predict_proba(x_train)
```

```python
# Ausgabe der Wahrscheinlichkeit für Testdatensatz, dass die Pixelwerte einer Ziffer zugeordnet wurden
y_predict_test_proba = logisticRegr.predict_proba(x_test)
```

```python
Log_Loss_Train = []
Log_Loss_Test = []
log_loss_train = log_loss(y_train, y_predict_train_proba)
log_loss_test = log_loss(y_test, y_predict_test_proba)

Log_Loss_Train.append(log_loss_train)
Log_Loss_Test.append(log_loss_test)
```

```python
type(log_loss_test)
```

```python
# Berechnung von log loss, je kleiner der Wert, desto besser die prediction
# log loss ist für Klassifizierungen ein Qualitätsmaß, wie der RMSE ein Qualitätsmaß für lineare Regression ist
pd_log_loss = pd.DataFrame()
pd_log_loss["log loss train"] = Log_Loss_Train
pd_log_loss["log loss test"] = Log_Loss_Test
pd_log_loss
```

```python
# Darstellung von Accuracy, Error und log loss in eine DataFrame
summary_acc_error_log_loss = pd.concat((pd_accuracy_score_train_test, pd_log_loss), axis = 1)
summary_acc_error_log_loss
```

### precision score für Traindaten

### recall score für Traindaten

### Berechnung des F1-scores für Traindaten

```python
f1_score(y_train, y_predict_train, average="weighted")
```

```python
# Berechnung von precision score,recall score und f1_score über eine Funktion für jede einzelne Ziffer
train_precision=[]
train_recall=[]
train_f1_score=[]

def precision_recall_f1(number):
    y_train_number = y_train==number
    y_predict_train_number = y_predict_train == number
    train_ps = precision_score(y_train_number, y_predict_train_number)
    train_rc = recall_score(y_train_number, y_predict_train_number)
    train_f1 = f1_score(y_train_number, y_predict_train_number)
    
    train_precision.append(train_ps)
    train_recall.append(train_rc)
    train_f1_score.append(train_f1)
    
    return train_precision, train_recall, train_f1_score
                 
```

```python
numbers = range(0,10)
for number in numbers:
    precision_recall_f1(number)
```

```python
summary_train_ps_rc_f1 = pd.DataFrame()
summary_train_ps_rc_f1["Precision"] = train_precision
summary_train_ps_rc_f1["Recall"] = train_recall
summary_train_ps_rc_f1["F1-Score"] = train_f1_score
summary_train_ps_rc_f1
```

```python
y_scores_train = logisticRegr.decision_function(x_train)
```

```python
# Anzeige nur der 0 in y_train
y_train_0 = y_train==0
```

```python
y_train_scores_0 = cross_val_predict(logisticRegr, x_train, y_train_0, cv=3, method="decision_function")
```

```python
precisions, recalls, thresholds = precision_recall_curve(y_train_0, y_train_scores_0)
```

```python
def plot_precision_recall_vs_thresholds(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1],"b--", label="Relevanz")
    plt.plot(thresholds, recalls [:-1], "g--", label="Sensitivität")
    plt.xlabel("Schwellenwert")
    plt.legend(loc="center left")
    plt.ylim([0,1])
    plt.title("Ziffer 0")

plot_precision_recall_vs_thresholds(precisions, recalls, thresholds)
plt.show()
```

#### Precision-Recall-Kurve für alle Ziffern

```python
def plot_precision_recall_vs_thresholds_alldigits(number):
    y_train_number = y_train==number
    y_train_scores_number = cross_val_predict(logisticRegr, x_train, y_train_number, cv=3, method="decision_function")
    precisions, recalls, thresholds = precision_recall_curve(y_train_number, y_train_scores_number)
    
    plt.figure(figsize=(20,8))
    plt.plot(thresholds, precisions[:-1],"b--", label="Relevanz")
    plt.plot(thresholds, recalls [:-1], "g--", label="Sensitivität")
    plt.xlabel("Schwellenwert")
    plt.legend(loc="center left")
    plt.ylim([0,1])
    plt.title(f'Ziffer: {number}')
    
    return plt.show()
```

```python
numbers = range(0,10)
for number in numbers:
    plot_precision_recall_vs_thresholds_alldigits(number)
```

## Erstellung einer ROC-Kurve, ROC = Receiver Operating Characteristic)

#### ROC-Kurve für die Ziffer 0

```python
fpr, tpr, tresholds = roc_curve(y_train_0, y_train_scores_0)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.figure(figsize=(20,8))
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0,1],[0,1], "k--")
    plt.axis([0,1,0,1])
    plt.xlabel("False-positive Rate (FPR)")
    plt.ylabel("True-positive Rate(TPR)")
    
plot_roc_curve(fpr,tpr)
plt.show()
```

#### ROC-Kurve für alle Ziffern für Train- und Testdatensatz

```python
def roc_curve_all_digits(number):
    fpr_train, tpr_train, tresholds_train = roc_curve(y_train, y_predict_train_proba[:,number], pos_label=number)
    fpr_test, tpr_test, tresholds_test = roc_curve(y_test, y_predict_test_proba[:,number], pos_label = number)
    
    plt.figure(figsize=(20,10))
    plt.plot(fpr_train, tpr_train, linewidth=2, color = "green")
    plt.plot(fpr_test, tpr_test, linewidth=2, color = "red")
    plt.plot([0,1],[0,1], "k--")
    plt.axis([0,1,0,1])
    plt.xlabel("False-positive Rate (FPR)")
    plt.ylabel("True-positive Rate(TPR)")
    plt.title(f'Ziffer: {number}')
    return plt.show()
```

```python
numbers = range(0,10)
for number in numbers:
    roc_curve_all_digits(number)
```

#### Berechnung des ROC-AUC-Scores für die Ziffer 0

```python
roc_auc_score(y_train_0, y_train_scores_0)
```

#### Berechnung des ROC-AUC-Scores für alle Ziffern

```python
train_roc_auc_scores =[]
test_roc_auc_scores = []
def roc_auc_score_alldigits(number):
    y_train_number = y_train==number
    y_test_number = y_test == number
    y_train_scores_number = cross_val_predict(logisticRegr, x_train, y_train_number, cv=3, method="decision_function")
    y_test_scores_number = cross_val_predict(logisticRegr, x_test, y_test_number, cv=3, method="decision_function")
    roc_auc_score_train_number= roc_auc_score(y_train_number, y_train_scores_number)
    roc_auc_score_test_number = roc_auc_score(y_test_number, y_test_scores_number)
    train_roc_auc_scores.append(roc_auc_score_train_number)
    test_roc_auc_scores.append(roc_auc_score_test_number)
    return train_roc_auc_scores, test_roc_auc_scores
```

```python
numbers = range(0,10)
for number in numbers:
    roc_auc_score_alldigits(number)
```

```python
summary_roc_auc_score = pd.DataFrame()
summary_roc_auc_score["Train"] = train_roc_auc_scores
summary_roc_auc_score["Test"] = test_roc_auc_scores
summary_roc_auc_score
```

## Fehleranalyse

#### Darstellung der Fehler mit einer Konfusionsmatrix

```python
y_train_predict = cross_val_predict(logisticRegr, x_train, y_train, cv=3)
conf_mx = confusion_matrix(y_train, y_train_predict)
conf_mx
```

#### graphische Darstellung einer Konfusionsmatrix

```python
plt.matshow(conf_mx, cmap=plt.cm.gray)
plt.show()
# die Ziffern 0 und 1, da diese Felder am hellsten sind, wurden am besten zueinander zugeordnet. 
# y-Achse=y-true, x-Achse=y_predict
# Fehler sind absolut
```

#### graphische Darstellung einer Konfusionsmatrix mit relativen Fehlern

```python
row_sums = conf_mx.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mx / row_sums
```

```python
np.fill_diagonal(norm_conf_mx, 0) # Diagonale werden mit Nullen aufgefüllt, um nur die Fehler zu betrachten
plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plt.show()
# folgendermaßen lesen: 
# x-Achse/Zeile = predict, eine vorhergesagte 9 wurde fälschlicherweise oft als eine 7 (helles Feld) klassifiziert
# eine predictete 3 wurde oft als 5 verwechselt
# Verwechselung:
# 0 und 1 wurden ganz gut vorhergesagt
# 3 mit 5
# 4 mit 2 und generell schlecht
# 5 mit 3 und 8
# 6 mit 5
# 7 mit 9
# 8 mit 3
# 9 mit 7
# größten Probleme: 3,4,5,6,9
```

```python
# Umwandlung von norm_conf_mx in ein Pandas DataFrame
pd_norm_conf_mx = pd.DataFrame(norm_conf_mx)
pd_norm_conf_mx
```

## Feature Engineering

### Verschiebung der Pixel, um nicht zentrierte Ziffern besser zu erkennen:
### pro Richtung um 2 Pixel verschieben

```python
# x_train_shift2u1r = shift(x_train,[2,1],cval=0)
# [unten,rechts]
#[-oben,-links]
# Pixel werden um 2 nach unten (2u) und um 1 Pixel nach rechts (1r) verschoben
# [-2,1] --> Pixel werden um 2 nach oben und um 1 nach rechts verschoben
# [2,-1 --> Pixel werden um 2 nach unten und um 1 nach links verschoben
```

#### Verschiebung der Pixel um 2 in jede Richtung(o=oben, r=rechts, u=unten, l=links) verschoben

```python
x_train_shift_2u = shift(x_train,[2,0],cval=0)
x_train_shift_2r = shift(x_train,[0,2], cval=0)
x_train_shift_2o = shift(x_train,[-2,0], cval=0)
x_train_shift_2l = shift(x_train,[0,-2], cval=0)
```

#### Erstellung eines Dataframes

```python
pd_x_train_shift_2u = pd.DataFrame(x_train_shift_2u)
pd_x_train_shift_2r = pd.DataFrame(x_train_shift_2r)
pd_x_train_shift_2o = pd.DataFrame(x_train_shift_2o)
pd_x_train_shift_2l = pd.DataFrame(x_train_shift_2l)
```

#### Vereinen von y_train mit verschobenenem DataFrame, damit veränderte Pixel das richtige label besitzen

```python
def y_add_shift_df(label, df_shifted):
    y_df_shifted = pd.concat((label, df_shifted), axis=1)
    return y_df_shifted
```

```python
y_add_shift_df(y_train, pd_x_train_shift_2u)
```
